## Question 3: 
Write a class named Marsupial, Now write a class named Kangaroo as a subclass of Marsupial that inherits all the
attributes of Marsupial 

```python
class Marsupial:
    
    def __init__(me):
        me.list=[]
        
    def put_in_pouch(me,add):
        me.list.append(add)
        
    def pouch_contents(me):
        print(me.list)
        
m=Marsupial()
m.put_in_pouch("Doll")
m.put_in_pouch("Firetruck")
m.put_in_pouch("Kitten")
m.pouch_contents()
```

```python
class kangaroo(Marsupial):
    
    def __init__(me,a,b):
#     a = 0, b = 0
        me.x=a
        me.y=b
        me.list=[]
        
# k=kangaroo(0,0)
#print(k)
        
    def __str__(me):
        return "I am a kangaroo located at coordinates ("+str(me.x)+","+str(me.y)+")"
    
    def jump(me,dx,dy):
        me.x+=dx
        me.y+=dy
        return "I am a kangaroo located at coordinates ("+str(me.x)+","+str(me.y)+")"
    
k=kangaroo(0,0)
print(k)
    
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

```python
from urllib.request import urlopen
# from bs4 import BeautifulSoup as soup
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    switcher = {
        "h1": "",
        "h2": " "*1,
        "h3": " "*2,
        "h4": " "*3,
        "h5": " "*4,
        "h6": " "*5
    }
    current_tag = ""
    heading_line = []

    def handle_starttag(self, tag, attrs):
        self.current_tag = tag

    def handle_data(self, data):
        if self.current_tag in ["h1", "h2", "h3", "h4", "h5", "h6"]:
            self.heading_line.append(self.switcher.get(self.current_tag, "")+data)

    def handle_endtag(self, tag):
        if self.current_tag == tag:
            self.current_tag = ""

if __name__ == "__main__":
    infile = open('C:/Users/Deepali/Downloads/w3c.html')
    content = infile.read()
    infile.close()
    hp = HeadingParser()
    hp.feed(content)
    for line in hp.heading_line:
        print(line)

```
